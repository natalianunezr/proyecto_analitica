```python
#Importamos las librerías a usar
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Importamos los datos como dataframe
df = pd.read_csv('processed.cleveland.csv', header=0)

#Convertimos datos a números
df = df.apply(pd.to_numeric, errors='coerce')
#Verificar importación
#print(df.head())

#En donde tengamos datos faltantes (?) ponemos 0

for column in df.columns:
    # Reemplazar valores "?" con 0 en la columna y fila respectivas
    df[column] = df[column].replace('?', 0)


#print(df.columns)

#Estadísticas descriptivas

#Medias
# Creo un diccionario con los nombres de las variables como claves y las medias como valores
medias_dict = {"age": df["age"].mean(),
               "trestbps": df["trestbps"].mean(),
               "chol": df["chol"].mean(),
               "thalach": df["thalach"].mean(),
               "oldpeak": df["oldpeak"].mean(),
               "ca": df["ca"].mean(),
               "num": df["num"].mean()}

# Creo un nuevo DataFrame a partir del diccionario de medias
medias = pd.DataFrame.from_dict(medias_dict, orient='index',columns=['media'])

print(medias)

#Contador de mujeres y hombres
conteo_sexo = df.groupby("sex")["age"].count()
num_hombres = conteo_sexo[1]
num_mujeres = conteo_sexo[0]
print("Número de hombres: ", num_hombres)
print("Número de mujeres: ", num_mujeres)

#Contador por tipo de dolor toráxico
# Creamos un diccionario que asocia cada nivel con el valor correspondiente de "cp"
niveles_dict = {1: 1, 2: 2, 3: 3, 4: 0}

# Creamos una nueva columna "nivel" que asigna el valor de "cp" correspondiente a cada nivel
df["nivel"] = df["cp"].map(niveles_dict)

# Usamos groupby y value_counts para contar el número de veces que aparece cada valor de "nivel"
conteo_niveles = df.groupby("nivel")["cp"].value_counts()

# Imprimimos los resultados
print(conteo_niveles)
```

```python
#Estadísticas para las categóricas:
#Proporciones para sex
# Contamos las veces que aparece cada valor en la variable categórica
counts_sex = df["sex"].value_counts()

# Calculamos las proporciones
proporciones_sex = counts_sex / len(df)

# Creamos una tabla con los resultados
tabla_proporciones_sex = pd.DataFrame({'Categoría': counts_sex.index, 'Cantidad': counts_sex.values, 'Proporción - Sex': proporciones_sex.values})

# Imprimimos la tabla
print(tabla_proporciones_sex)

#Proporciones para cp
# Contamos las veces que aparece cada valor en la variable categórica
counts_cp = df["cp"].value_counts()

# Calculamos las proporciones
proporciones_cp = counts_cp / len(df)

# Creamos una tabla con los resultados
tabla_proporciones_cp = pd.DataFrame({'Categoría': counts_cp.index, 'Cantidad': counts_cp.values, 'Proporción - Cp': proporciones_cp.values})

# Imprimimos la tabla
print(tabla_proporciones_cp)

#Proporciones para fbs
# Contamos las veces que aparece cada valor en la variable categórica
counts_fbs = df["fbs"].value_counts()

# Calculamos las proporciones
proporciones_fbs = counts_fbs / len(df)

# Creamos una tabla con los resultados
tabla_proporciones_fbs = pd.DataFrame({'Categoría': counts_fbs.index, 'Cantidad': counts_fbs.values, 'Proporción - Fbs': proporciones_fbs.values})

# Imprimimos la tabla
print(tabla_proporciones_fbs)

#Proporciones para restecg
# Contamos las veces que aparece cada valor en la variable categórica
counts_restecg = df["restecg"].value_counts()

# Calculamos las proporciones
proporciones_restecg = counts_restecg / len(df)

# Creamos una tabla con los resultados
tabla_proporciones_restecg = pd.DataFrame({'Categoría': counts_restecg.index, 'Cantidad': counts_restecg.values, 'Proporción - Restecg': proporciones_restecg.values})

# Imprimimos la tabla
print(tabla_proporciones_restecg)

#Proporciones para exang
# Contamos las veces que aparece cada valor en la variable categórica
counts_exang = df["exang"].value_counts()

# Calculamos las proporciones
proporciones_exang = counts_exang / len(df)

# Creamos una tabla con los resultados
tabla_proporciones_exang = pd.DataFrame({'Categoría': counts_exang.index, 'Cantidad': counts_exang.values, 'Proporción - Exang': proporciones_exang.values})

# Imprimimos la tabla
print(tabla_proporciones_exang)

#Proporciones para slope

#Contamos las veces que aparece cada valor en la variable categórica
counts_slope = df["slope"].value_counts()

#Calculamos las proporciones
proporciones_slope = counts_slope / len(df)

#Creamos una tabla con los resultados
tabla_proporciones_slope = pd.DataFrame({'Categoría': counts_slope.index, 'Cantidad': counts_slope.values, 'Proporción - Slope': proporciones_slope.values})

#Imprimimos la tabla
print(tabla_proporciones_slope)

#Proporciones para thal
# Contamos las veces que aparece cada valor en la variable categórica
counts_thal = df["thal"].value_counts()

# Calculamos las proporciones
proporciones_thal = counts_thal / len(df)

# Creamos una tabla con los resultados
tabla_proporciones_thal = pd.DataFrame({'Categoría': counts_thal.index, 'Cantidad': counts_thal.values, 'Proporción - Thal': proporciones_thal.values})

# Imprimimos la tabla
print(tabla_proporciones_thal)
```

```python
#Diagramas de dispersión entre categóricas:
#PREGUNTA: ¿Hay una relación entre la edad del paciente y su nivel de colesterol sérico que difiere por género?
# Cargar los datos en un DataFrame de Pandas para este análisis
data_csv = pd.read_csv('processed.cleveland.csv')

# Agrupar los datos por género
grouped_data_csv = data_csv.groupby('sex')

# Crear un gráfico de dispersión para cada grupo
fig, axs = plt.subplots(nrows=2, sharex=True)
for ax, (gender, group) in zip(axs, grouped_data_csv):
    ax.scatter(group['age'], group['chol'], label=gender)
    ax.legend()
    ax.set_xlabel('Edad')
    ax.set_ylabel('Colesterol sérico')
    
plt.show()
```

```python
# Crear un gráfico de dispersión de la edad y el colesterol sérico
plt.scatter(df['age'], df['chol'], color='green')

# Agregar etiquetas a los ejes x e y
plt.xlabel('Edad')
plt.ylabel('Colesterol sérico')

# Calcular el coeficiente de correlación
corr = df['age'].corr(df['chol'])
plt.text(20, 600, f'Coeficiente de correlación: {corr:.2f}')

# Agregar leyenda al gráfico
plt.legend(['Edad vs. Colesterol sérico'])

# Cambiar el color de fondo del gráfico
plt.rcParams['axes.facecolor'] = 'lightgrey'

# Mostrar el gráfico
plt.show()
```
